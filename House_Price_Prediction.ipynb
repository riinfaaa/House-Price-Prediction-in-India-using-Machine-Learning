```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/MACHINE LEARNING/Housing.csv")
```

```python
print(data)
```

```python
data.isna().sum()
```

```python
data.dtypes
```

```python

le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
le4 = LabelEncoder()
le5 = LabelEncoder()
le6 = LabelEncoder()
le7 = LabelEncoder()
```

```python
data['mainroad'] = le1.fit_transform(data['mainroad'])
data['guestroom'] = le2.fit_transform(data['guestroom'])
data['basement'] = le3.fit_transform(data['basement'])
data['hotwaterheating'] = le4.fit_transform(data['hotwaterheating'])
data['airconditioning'] = le5.fit_transform(data['airconditioning'])
data['prefarea'] = le6.fit_transform(data['prefarea'])
data['furnishingstatus'] = le7.fit_transform(data['furnishingstatus'])

print(data)
```

```python
data.dtypes
```

```python
x = data.drop('price', axis =1 )
y = data['price']
model = LinearRegression()
model.fit(x,y)
```

```python
model.coef_
```

c value:

```python
model.intercept_
```

```python
ypred = model.predict(x)
pd.DataFrame({'true y value': y, 'predicted y value': ypred})
```

```python
f,ax = plt.subplots(figsize=(6,6))
ax.scatter(y, ypred)

plt.xlabel('target y values')
plt.ylabel('predicted y values')

ax.plot([0,1],[0,1],transform = ax.transAxes, c ='red')
```

```python
cor = data.corr()
cor
```

```python
import seaborn as sns
sns.heatmap(cor, annot=True, cmap = 'Greys')
```

checking the correctness of model:

```python
model.score(x,y)
```

```python
ypred = model1.predict(x)
plt.scatter(y, ypred)
```
